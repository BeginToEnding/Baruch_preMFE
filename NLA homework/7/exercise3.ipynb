```python
import pandas as pd
import sympy as sp
import numpy as np
```

```python
financial_data = pd.read_excel("financials2016-8.xlsx")
financial_data
```

```python
weekly_returns = financial_data.set_index('Date').pct_change(-1) * 100
weekly_returns.index = weekly_returns.index.date
weekly_returns.head()
```

```python
print(weekly_returns.to_latex(float_format="%.2f"))
```

```python
y = weekly_returns['JPM'].dropna()
X = weekly_returns.drop(columns=['JPM']).loc[y.index]

# Display X and y to confirm their structure
print(X.head().to_latex(float_format="%.2f"))
print(y.head().to_latex(float_format="%.2f"))
```

```python

# Prepare X and y as numpy arrays for regression calculation
X_np = np.array(X.dropna())
y_np = np.array(y[X.index.intersection(y.index)])

# Add an intercept term with a column of ones to X
X_with_intercept = np.hstack([np.ones((X_np.shape[0], 1)), X_np])

# Calculate the beta coefficients using the Normal Equation
beta = np.linalg.inv(X_with_intercept.T @ X_with_intercept) @ X_with_intercept.T @ y_np

# Display the beta coefficients
beta
```

```python
sp.Matrix(beta.round(3))
```

```python
# Predicted values of y based on the regression model
y_pred = X_with_intercept @ beta

# Residuals (errors) calculation
epsilon = y_np - y_pred

np.linalg.norm(epsilon)
```

```python
raw_financial_data = financial_data
y_raw = raw_financial_data['JPM'].dropna()
X_raw = raw_financial_data.drop(columns=['JPM', 'Date']).loc[y_raw.index]
X_raw_with_intercept = np.hstack([np.ones((X_raw.shape[0], 1)), X_raw.values])
beta_raw = np.linalg.inv(X_raw_with_intercept.T @ X_raw_with_intercept) @ X_raw_with_intercept.T @ y_raw.values

# Display the beta coefficients for the raw data fitting
sp.Matrix(beta_raw.round(3))
```

```python
y_pred = X_raw_with_intercept @ beta_raw
epsilon = y_raw - y_pred
np.linalg.norm(epsilon)
```
