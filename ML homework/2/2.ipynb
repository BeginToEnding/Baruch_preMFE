```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters
n = 5  # Number of trials
p = 0.5  # Probability of success (e.g., heads)

# Possible values for X
x = np.arange(0, n + 1)

# Probability mass function (pmf)
pmf = binom.pmf(x, n, p)

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.xlabel('$X$ (Number of heads in 5 flips)', fontsize=12)
plt.ylabel('$P(X = i)$', fontsize=12)
plt.title('Probability Mass Function of $X \\sim \\mathrm{Binomial}(5, 0.5)$')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Saving as PDF
plt.savefig("binomial_pmf_plot.pdf")

plt.show()
```
